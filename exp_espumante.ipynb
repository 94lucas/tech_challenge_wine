```python
dados_espumante.set_index('País', inplace=True)
```

```python
def separar_valor_kg(dados):
    coluna_valor = dados.columns[dados.columns.str.contains('\.')]
    coluna_kg = dados.columns[~dados.columns.str.contains('\.')]
    dados_espumante_valor = dados[coluna_valor]
    dados_espumante_valor.index = dados.index
    dados_espumante_kg = dados[coluna_kg]
    return dados_espumante_valor, dados_espumante_kg

```

```python
df_espumante_valor, df_espumante_kg = separar_valor_kg(dados_espumante)
```

```python
display(df_espumante_kg.tail(2))
display(df_espumante_valor.tail(2))
```

```python
df_espumante_kg.drop(df_espumante_kg.index[-1],axis=0, inplace=True)
df_espumante_valor.drop(df_espumante_valor.index[-1],axis=0, inplace=True)
```

```python
def dados_aberto(dados, label = 'coluna'):
    if dados.index is 'País':
        dados = dados.set_index('País').T.reset_index()
        dados_aberto = pd.melt(dados, id_vars=['index'],var_name='pais', value_name=f'{label}')
    else:
        dados = dados.T.reset_index()
        dados_aberto = pd.melt(dados, id_vars=['index'],var_name='pais', value_name=f'{label}')
    return dados_aberto
```

```python
df_espumante_valor_aberto = dados_aberto(df_espumante_valor, 'us$')
df_espumante_valor_aberto.rename(columns={'index':'ano'}, inplace=True)
```

```python
df_espumante_kg_aberto = dados_aberto(df_espumante_kg, 'kg')
df_espumante_kg_aberto.rename(columns={'index':'ano'}, inplace=True)
df_espumante_kg_aberto.head()
```

```python
df_espumante_valor_aberto['ano'] = df_espumante_valor_aberto['ano'].map(lambda x: str(x).replace('.1',''))
```
